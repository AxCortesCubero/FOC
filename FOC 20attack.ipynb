```python
## calculating the consensus pledge (with bug fixed)
consensuspledge=3.436e10*0.3*cil_df['circ_supply']*(0.3/cil_df['network_baseline']+0.7/cil_df['network_QAP'])
```

```python

fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(cil_df['date'], consensuspledge)

plt.xlabel("time")
plt.ylabel("consensus pledge")


plt.show()
```

```python
storagepledge=np.zeros(cil_df['days'].iloc[-1]-cil_df['days'].iloc[0])
N=cil_df['days'].iloc[-1]-cil_df['days'].iloc[0]

for n in range(N):
    storagepledge[n]=3.436e+10*np.sum(cil_df['day_network_reward'][n:n+20]/cil_df['network_QAP'][n:n+20])/20
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(cil_df['date'][0:N], storagepledge)

plt.xlabel("time")
plt.ylabel("consensus pledge")


plt.show()
```

```python
Powerchangepercentage=cil_df['network_QAP'][N]/cil_df['network_QAP'][0]
Pledgechangepercentage=(consensuspledge[N-1]+storagepledge[N-1])/(consensuspledge[0]+storagepledge[0])
```
